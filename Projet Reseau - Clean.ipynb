```python
LIGNES = np.where(CSV["LINK_SENTIMENT"] == 1)
LIGNES = list(LIGNES[0])
BAD_CSV = CSV.drop(LIGNES)
good = len(LIGNES)
bad = len(BAD_CSV)
ratio = bad/good

rd.shuffle(LIGNES)
LIGNES = LIGNES[:int(len(LIGNES)/1.02)]
NEW_CSV = CSV.drop(LIGNES)


BAD_LIGNES = list(BAD_CSV.index)
BAD_LIGNES = BAD_LIGNES[:int(ratio*(1-good+len(LIGNES)))]
NEW_CSV = NEW_CSV.drop(BAD_LIGNES)
NEW_CSV
```
